```python
#imports
import numpy as np
import math
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
from PIL import Image
import os

#tensorboard
%load_ext tensorboard
from tensorflow import summary
import tensorflow as tf

#pytorch
import torch
import torchvision
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

#directory for saving tensorboard data and model
if not os.path.exists('./models'):
    os.makedirs('./models')
```

```python
#set seed
%env CUBLAS_WORKSPACE_CONFIG=:16:8
torch.backends.cudnn.deterministic = True
torch.set_deterministic(True)
torch.manual_seed(19)
torch.cuda.manual_seed(19)
np.random.seed(19)
```

```python
#prepare image tensors and labels
class facemask():
    
    def __init__(self, path):
        self.path = path
        
    def __len__(self):
        return len(os.listdir(self.path))

    def __getitem__(self, idx):
        
        Img=Image.open(self.path+os.listdir(self.path)[idx])
        to_tensor=torchvision.transforms.ToTensor()
        
        data=dict()
        data['img']=to_tensor(Img)
        
        if os.listdir(self.path)[idx][-5]=='k':
            data['class']= "mask"
            data['label']= 1
        else:    
            data['class']= "no mask"
            data['label']= 0  
            
        return data
```

```python
#prepare dataset

dataset= facemask(path)
L=len(dataset)

#split in train, test, validation
train_set, val_set, test_set = \
torch.utils.data.random_split(dataset,[int(L*0.8),int(L*0.1),math.ceil(L*0.1)])

print(len(train_set),'+', len(val_set),'+', len(test_set),'=',L)

#dataloader
batch_size=128

dataset_dl=DataLoader(dataset, batch_size=batch_size, shuffle=True)
train_dl = DataLoader(train_set, batch_size=batch_size, shuffle=True)
val_dl = DataLoader(val_set, batch_size=1, shuffle=False)
test_dl = DataLoader(test_set, batch_size=1, shuffle=False)
```

```python
#view the dataloader
sample=next(iter(train_dl))
print(sample['img'].shape) # batch_size, channels, height, width
print(sample['label'])# 1 or 0
```

```python
#visualization
fig, ax = plt.subplots(1, 7, figsize=(20, 10))
for i in range(7):
    ax[i].imshow(sample['img'][i].permute(1, 2, 0))
    ax[i].set_title(sample['class'][i])
```

```python
#CNNs
class Model(nn.Module):
    def __init__(self, height=218, width=178, dim_output=2):
        super(Model, self).__init__()
        
        self.H=height
        self.W=width
        self.output=dim_output
        
        self.max_pool = nn.MaxPool2d(kernel_size=2, stride=2)
        self.activation = nn.ReLU()
        self.dropout= nn.Dropout(p=0.5) 
        self.bn1= nn.BatchNorm2d(32) 
        self.bn2= nn.BatchNorm2d(64)
        self.bn3= nn.BatchNorm2d(128)
        
        #convolutions
        self.conv1 = nn.Conv2d( 3,  32,kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv2d(32,  64,kernel_size=3, stride=1, padding=1)
        self.conv3 = nn.Conv2d(64, 128,kernel_size=3, stride=1, padding=1)
        
        #fully connected
        self.fc1 = nn.Linear(128*(self.H//8)*(self.W//8),128) 
        self.fc2 = nn.Linear(128,64)
        self.fc3 = nn.Linear(64,self.output)
        
        #sequential
        self.layer1=nn.Sequential(self.conv1, self.bn1, self.max_pool, self.activation)
        self.layer2=nn.Sequential(self.conv2, self.bn2, self.max_pool, self.activation)
        self.layer3=nn.Sequential(self.conv3, self.bn3, self.max_pool, self.activation)
        self.layer4=nn.Sequential(self.fc1, self.activation, self.dropout)
        self.layer5=nn.Sequential(self.fc2, self.activation, self.dropout)
        
    def forward(self, img):
        
        out=self.layer3(self.layer2(self.layer1(img)))
        
        ###flattening
        batch_size =img.shape[0]
        out=out.view(batch_size, -1)
        
        out=self.fc3(self.layer5(self.layer4(out)))
       
        return out           
```

```python
#model and optimizer
model = Model()
model = model.to(device) 

optimizer = optim.Adam(model.parameters(), lr=1e-5, weight_decay=1e-5)

#print(model)
#sample_out=model(sample['img'].to(device)) 
#sample_out.shape
```

```python
#train function
def train(model, optimizer, sample):
    model.train()

    criterion = nn.CrossEntropyLoss()
    
    # reset optimizer's gradient
    optimizer.zero_grad()

    # define input and output
    img = sample['img'].float().to(device) 
    label = sample['label'].long().to(device)
  
    # get the prediction
    pred = model(img) 

    pred_loss = criterion(pred, label)
    
    # regularization loss
    reg_loss= 0.0
    for p in model.parameters():
        if p.requires_grad: # if p is a trainable parameter
            reg_loss+=torch.norm(p,2)

    total_loss= pred_loss + 1e-6 * reg_loss
    
    num_correct = sum(torch.argmax(pred, dim=1) == label)
    
    #one step of training
    total_loss.backward()
    optimizer.step()

    return total_loss.item(), num_correct.item()
```

```python
#test function
def test(model, sample): 
    model.eval()

    criterion = nn.CrossEntropyLoss()
    
    with torch.no_grad(): 
        
        img = sample['img'].float().to(device)
        label = sample['label'].long().to(device)
       
        # get the prediction
        pred = model(img)
        
        pred_loss = criterion(pred, label)
        
        # regularization loss
        reg_loss= 0.0
        for p in model.parameters():
            if p.requires_grad: # if p is a trainable parameter
                reg_loss+=torch.norm(p,2)

        total_loss= pred_loss + 1e-6 * reg_loss
        
        num_correct = sum(torch.argmax(pred, dim=1) == label)

    return total_loss.item(), num_correct.item()
```

```python
#prepare tensorboard
train_log_dir = './models/runs/train'
train_summary_writer = summary.create_file_writer(train_log_dir)
val_log_dir = './models/runs/validate'
val_summary_writer = summary.create_file_writer(val_log_dir)
```

```python
##################################start training and validating#####################################################
max_epoch = 10
tmp_path = './models/temp.pth'
max_accu = -1

for epoch in tqdm(range(max_epoch)):        
    ###Train Phase
    
    # Initialize Loss and Accuracy
    train_loss = 0.0
    train_accu = 0.0

    #########TEMPORARY################################
    # Load the saved MODEL AND OPTIMIZER after evaluation.
    if epoch > 0:
        checkpoint = torch.load(tmp_path)
        model.load_state_dict(checkpoint['model_state_dict'])
        optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
        
    ##################################################

    # Iterate over the train_dataloader
    with tqdm(total=len(train_dl)) as pbar:
        for sample in train_dl:            
            curr_loss, num_correct = train(model, optimizer, sample)
            train_loss += curr_loss / len(train_dl) 
            train_accu += num_correct / len(train_set)
            pbar.update(1)

    ### Write the current loss and accuracy to the Tensorboard
    with train_summary_writer.as_default():
        tf.summary.scalar('loss', train_loss, step=epoch)                
        tf.summary.scalar('accuracy', train_accu, step=epoch)                

    ###############TEMPORARY#######################################
    # save the model and optimizer's information before the evaulation
    checkpoint = {
         'model_state_dict': model.state_dict(),
         'optimizer_state_dict': optimizer.state_dict()}
    torch.save(checkpoint, tmp_path)
    ###################################################################
    
    ### Validate Phase
    # Initialize Loss and Accuracy
    val_loss = 0.0
    val_accu = 0.0

    # Iterate over the test_dataloader
    with tqdm(total=len(val_dl)) as pbar:
        for idx, sample in enumerate(val_dl): 
            curr_loss, num_correct = test(model, sample)
            val_loss += curr_loss / len(val_dl)
            val_accu += num_correct / len(val_set)
            pbar.update(1)

    ### Write the current loss and accuracy to the Tensorboard
    with val_summary_writer.as_default():
        tf.summary.scalar('loss', val_loss, step=epoch)
        tf.summary.scalar('accuracy', val_accu, step=epoch)
        
#    max_accu = max(val_accu, max_accu)
#    if max_accu == val_accu:
#        # Save your best model to the checkpoint
#        torch.save(checkpoint, os.path.join('./models', 'best.pth'))
        
    print(train_accu, val_accu)   

#remove temp 
os.remove(tmp_path)
```

```python
#report results on test set
test_loss = 0.0
test_accu = 0.0
with tqdm(total=len(test_dl)) as pbar:
    for idx, sample in enumerate(test_dl): 
        curr_loss, num_correct = test(model, sample)
        test_loss += curr_loss / len(test_dl)
        test_accu += num_correct / len(test_set)
        pbar.update(1)
print(test_accu)
```
